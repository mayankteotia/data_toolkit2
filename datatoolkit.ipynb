```python
#Q12->
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Step 1: Create a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.normal(loc=20, scale=5, size=100)  # Normal distribution around 20 degrees
humidity = np.random.normal(loc=60, scale=10, size=100)    # Normal distribution around 60%
time_series_df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Plot Temperature and Humidity on the same plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature
ax1.plot(time_series_df['Date'], time_series_df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(time_series_df['Date'], time_series_df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set title
plt.title('Temperature and Humidity Over Time')
plt.show()

# Step 3: Create a NumPy array with samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 4: Plot histogram with PDF overlay
plt.figure(figsize=(12, 6))
sns.histplot(data, bins=30, kde=False, stat='density', color='skyblue', label='Histogram', alpha=0.6)
# Plot PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')
plt.title('Histogram with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# Step 5: Create a scatter plot of two random arrays
x_random = np.random.rand(100)
y_random = np.random.rand(100)

# Determine the quadrant based on the position of points relative to the origin
colors = ['red' if x > 0.5 and y > 0.5 else 'blue' if x <= 0.5 and y > 0.5 else
          'green' if x <= 0.5 and y <= 0.5 else 'orange' for x, y in zip(x_random, y_random)]

# Create scatter plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(x_random, y_random, c=colors, alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Create a legend
red_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 1', markerfacecolor='red', markersize=10)
blue_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 2', markerfacecolor='blue', markersize=10)
green_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 3', markerfacecolor='green', markersize=10)
orange_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 4', markerfacecolor='orange', markersize=10)
plt.legend(handles=[red_patch, blue_patch, green_patch, orange_patch])
plt.grid()
plt.show()
```

```python
#13->
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
mean = 0        # Mean of the normal distribution
std_dev = 1     # Standard deviation of the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Step 2: Plot a histogram of the data
plt.figure(figsize=(12, 6))
sns.histplot(data, bins=30, kde=False, stat='density', color='skyblue', label='Histogram', alpha=0.6)

# Step 3: Overlay the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=mean, scale=std_dev)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Customize the plot
plt.title('Histogram with PDF Overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
#Q14->
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate two random arrays
np.random.seed(0)  # For reproducibility
x_random = np.random.rand(100) * 10  # Random values from 0 to 10
y_random = np.random.rand(100) * 10  # Random values from 0 to 10

# Step 2: Determine the quadrant for each point
colors = []
for x, y in zip(x_random, y_random):
    if x >= 5 and y >= 5:
        colors.append('red')     # Quadrant 1
    elif x < 5 and y >= 5:
        colors.append('blue')    # Quadrant 2
    elif x < 5 and y < 5:
        colors.append('green')   # Quadrant 3
    else:
        colors.append('orange')   # Quadrant 4

# Step 3: Create a scatter plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(x_random, y_random, c=colors, alpha=0.7)

# Step 4: Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Step 5: Create a legend
red_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 1', markerfacecolor='red', markersize=10)
blue_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 2', markerfacecolor='blue', markersize=10)
green_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 3', markerfacecolor='green', markersize=10)
orange_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 4', markerfacecolor='orange', markersize=10)
plt.legend(handles=[red_patch, blue_patch, green_patch, orange_patch])

plt.grid()
plt.show()
```

```python
#Q14->
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Sample data
data = {
    'country': ["USA", "China", "India", "Japan", "Germany"],
    'gdp': [21.4, 14.3, 2.9, 5.1, 3.8],  # GDP in trillion USD
    'population': [331, 1439, 1380, 126, 83],  # Population in millions
    'size': [20, 60, 55, 15, 10]  # Bubble sizes proportional to population
}

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create the figure
p = figure(title="GDP vs Population Bubble Chart", 
           x_axis_label='GDP (in Trillion USD)', y_axis_label='Population (in Millions)',
           plot_width=700, plot_height=400, tools="")

# Add bubbles to the plot
p.circle(x='gdp', y='population', size='size', source=source, 
         fill_color=linear_cmap('population', Viridis256, min(data['population']), max(data['population'])),
         fill_alpha=0.7, line_color="white", legend_field='country')

# Display the plot inline
output_notebook()
show(p)
```

```python
#Q15->
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio

# Generate x and y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Compute z values for the function z = sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the surface plot
surface = go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')

# Configure the layout
layout = go.Layout(title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))", 
                   scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))

# Create the figure and display it
fig = go.Figure(data=[surface], layout=layout)
pio.show(fig)
```

```python
#Q18->
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral5

# Data
fruits = ["Apples", "Bananas", "Grapes", "Oranges", "Peaches"]
counts = [10, 15, 7, 12, 5]

# Create ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the bar chart
p = figure(x_range=fruits, title="Fruit Counts", 
           x_axis_label='Fruit', y_axis_label='Count', 
           plot_width=700, plot_height=400)

p.vbar(x='fruits', top='counts', width=0.6, source=source, 
       legend_field="fruits", line_color='white', 
       fill_color=factor_cmap('fruits', palette=Spectral5, factors=fruits))

# Display the plot inline
output_notebook()
show(p)
```

```python
#Q20->
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256
from bokeh.io import output_file

# Data setup
days = ["Mon", "Tue", "Wed"]
times = ["Morning", "Afternoon", "Evening"]
values = [3, 8, 5, 7, 2, 9, 4, 6, 1]

# Create a ColumnDataSource with the data
source = ColumnDataSource(data=dict(day=["Mon", "Mon", "Mon", 
                                         "Tue", "Tue", "Tue", 
                                         "Wed", "Wed", "Wed"],
                                    time=["Morning", "Afternoon", "Evening"] * 3,
                                    value=values))

# Create a color mapper based on the value
mapper = LinearColorMapper(palette=Viridis256, low=min(values), high=max(values))

# Create the figure
p = figure(title="Heatmap of Values", 
           x_range=days, y_range=list(reversed(times)), 
           plot_width=700, plot_height=400,
           toolbar_location=None, tools="")

# Add the rectangles representing each value
p.rect(x="day", y="time", width=1, height=1, source=source,
       line_color=None, fill_color={'field': 'value', 'transform': mapper})

# Add a color bar to the side
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Display the plot inline
output_notebook()
show(p)
```
